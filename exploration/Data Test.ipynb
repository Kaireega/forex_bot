```python

# Load the .pkl file

df = pd.read_pickle("../data/USD_JPY_M15.pkl")

# Export to whole set as JSON
json_path = "USD_JPY_M15"
# df.to_json(json_path, orient="records", lines=True)
# print(f"JSON data saved to {json_path}")

# Export to chunks as JSON
chunk_size = 1000  # Number of rows per chunk
for i in range(0, len(df), chunk_size):
    chunk = df.iloc[i:i + chunk_size]
    chunk.to_json(f" USD_JPY_M15 chunk_{i // chunk_size + 1}.json", orient="records", lines=True)
    print(f"JSON data saved to {json_path}_{i // chunk_size + 1}.json")
```
