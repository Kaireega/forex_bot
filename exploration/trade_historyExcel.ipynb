```python
import sys
sys.path.append("../")

import pandas as pd
import plotly.graph_objects as go
import datetime as dt
from technicals.indicators import BollingerBands, ATR, RSI, MACD  # Ensure these are imported
from models.trade_settings import TradeSettings
import json
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
```

```python
pair = "EUR_USD"
granularity = "H4"
df = pd.read_pickle(f"../data/{pair}_{granularity}.pkl")
```

```python
with open("../bot/settings.json", "r") as f:
    data = json.loads(f.read())

# Extract trade settings and risk
trade_settings = {k: TradeSettings(v, k) for k, v in data['pairs'].items()}
trade_risk = data['trade_risk']

# Example: Process for one pair (e.g., "USD_JPY")
pair = "USD_JPY"
settings = trade_settings[pair]
df_plot = pd.DataFrame(df)

# Apply indicators for the specific pair
df_plot = BollingerBands(df_plot, settings.n_ma, settings.n_std)
df_plot = ATR(df_plot, settings.atr_period)
df_plot = RSI(df_plot, settings.rsi_period)
df_plot = MACD(df_plot)
df_plot['BB_MA'] = df_plot['BB_MA'].fillna(method='ffill')
df_plot['BB_UP'] = df_plot['BB_UP'].fillna(method='ffill')
df_plot['BB_LW'] = df_plot['BB_LW'].fillna(method='ffill')
df_plot['ATR_14'] = df_plot['ATR_14'].fillna(method='ffill')
df_plot['RSI_14'] = df_plot['RSI_14'].fillna(method='ffill')
df_plot['MACD'] = df_plot['MACD'].fillna(method='ffill')
df_plot['SIGNAL_MD'] = df_plot['SIGNAL_MD'].fillna(method='ffill')
df_plot['HIST'] = df_plot['HIST'].fillna(method='ffill')



print(df_plot)
```
